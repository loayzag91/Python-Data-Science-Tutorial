```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

---

## `.barplot()`

To aggregate the categorical data based off some function, mean is used by default.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

To change the function used to aggregate the category, you can pass the **`estimator`** parameter.

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

---

## `.countplot()`

A bar plot where the estimator is explicitly set to count the number of occurences.

```python
sns.countplot(x='sex',data=tips)
```

---

## `.boxplot()`

A box plot (or box-and-whisker plot) shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

Each category can be further divided by passing the **`hue`** parameter.

Here, we separate the total bill for each day by whether or not the customer was a smoker.

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

---

## `.violinplot()`

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution on its side.

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

`.violinplot()` also accepts a **`hue`** parameter, but unlike the box plot, the violin plot can display the additional information as another graph or as the other half of the same graph.

These two alternatives are displayed below:

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

---

## `.stripplot()`

A scatterplot where one variable is categorical.

The **`hue`** and **`dodge`** parameters can manipulate the plots in similar ways as we have already seen.

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

---

## `.swarmplot()`

Plots a strip plot in the format of a violin plot.

*Note: avoid using swarm plots for large datasets. These plots can be computationally intensive and difficult to read for larger datasets.*

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

A swarm plot can be placed over a violin plot for some exploratory data analysis to help you immediately understand your data.

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

---

## `.catplot()`

You can access any categorical plot with `.catplot()` and through use of the **`kind`** parameter.

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```
